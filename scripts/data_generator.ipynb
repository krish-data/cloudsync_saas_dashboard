```python
import pandas as pd
import numpy as np
import random
import os
from datetime import datetime, timedelta

print("🚀 Booting up CloudSync Pro Data Generator...")

# 1. Setup Parameters
num_customers = 5000
np.random.seed(42)
random.seed(42)

# Date Range (Last 2 years)
start_date = datetime(2024, 1, 1)
end_date = datetime(2026, 3, 31)

# Generate IDs and Industries
customer_ids = [f"CUST-{str(i).zfill(5)}" for i in range(1, num_customers + 1)]
industries = random.choices(['Tech', 'Finance', 'Healthcare', 'Retail', 'Education'], weights=[40, 20, 15, 15, 10], k=num_customers)
signup_dates = [start_date + timedelta(days=random.randint(0, 800)) for _ in range(num_customers)]

# 2. Build the Business Logic (The Funnel & Churn)
tiers, mrr, churn_status, churn_dates, support_tickets, days_active = [], [], [], [], [], []

for i in range(num_customers):
    tier = random.choices(['Basic', 'Pro', 'Enterprise'], weights=[50, 35, 15])[0]
    tiers.append(tier)
    
    if tier == 'Basic':
        mrr.append(19)
        tickets = random.randint(0, 2)
        active = random.randint(5, 20)
        churn_prob = 0.10
    elif tier == 'Pro':
        mrr.append(49)
        # THE STORY: Pro users are experiencing a bug! High tickets = High Churn
        tickets = random.choices([1, 2, 5, 8], weights=[30, 30, 20, 20])[0] 
        active = random.randint(10, 25)
        churn_prob = 0.45 if tickets >= 5 else 0.15 
    else: # Enterprise
        mrr.append(199)
        tickets = random.randint(0, 3)
        active = random.randint(20, 30)
        churn_prob = 0.05
        
    support_tickets.append(tickets)
    days_active.append(active)
    
    is_churned = random.random() < churn_prob
    if is_churned:
        churn_status.append('Churned')
        churn_date = signup_dates[i] + timedelta(days=random.randint(30, 180))
        churn_dates.append(min(churn_date, end_date).strftime('%Y-%m-%d'))
        days_active[i] = random.randint(0, 5) 
    else:
        churn_status.append('Active')
        churn_dates.append(None)

# 3. Create the DataFrames
customers_df = pd.DataFrame({
    'customer_id': customer_ids, 'industry': industries, 'signup_date': [d.strftime('%Y-%m-%d') for d in signup_dates],
    'subscription_tier': tiers, 'mrr': mrr, 'churn_status': churn_status, 'churn_date': churn_dates
})

engagement_df = pd.DataFrame({
    'customer_id': customer_ids, 'days_active_last_30': days_active,
    'support_tickets_last_30': support_tickets, 'storage_used_gb': np.random.normal(loc=50, scale=15, size=num_customers).round(1)
})

# 4. Export to the 'data' folder safely
# This ensures it looks for the data folder one level up from the scripts folder
os.makedirs('../data', exist_ok=True)
customers_df.to_csv('../data/saas_customers.csv', index=False)
engagement_df.to_csv('../data/saas_engagement.csv', index=False)

print("✅ SUCCESS: Generated data for 5000 customers.")
print("💾 Saved 'saas_customers.csv' and 'saas_engagement.csv' into your 'data' folder.")
```
